```python
# import libraries
import pandas as pd
import plotly.express as px
```

## Read in Dataframes

```python
# read in cleaned overall dataframe
cleaned_candela_df = pd.read_csv('/home/ekmys/PycharmProjects/IMT542_FinalProject_Stelter/data/csv_files/cleaned_candela_df.csv')

# read in exploded dataframes
df_characters = pd.read_csv('/home/ekmys/PycharmProjects/IMT542_FinalProject_Stelter/data/csv_files/character_df_exploded.csv')
df_relationships = pd.read_csv('/home/ekmys/PycharmProjects/IMT542_FinalProject_Stelter/data/csv_files/relationships_df_exploded.csv')
df_fandom = pd.read_csv('/home/ekmys/PycharmProjects/IMT542_FinalProject_Stelter/data/csv_files/fandom_df_exploded.csv')
df_warnings = pd.read_csv('/home/ekmys/PycharmProjects/IMT542_FinalProject_Stelter/data/csv_files/warnings_df_exploded.csv')
df_additional = pd.read_csv('/home/ekmys/PycharmProjects/IMT542_FinalProject_Stelter/data/csv_files/additional_tags_df_exploded.csv')
df_categories = pd.read_csv('/home/ekmys/PycharmProjects/IMT542_FinalProject_Stelter/data/csv_files/categories_df_exploded.csv')
```

## Bar Charts: "Most Commonly Tagged"

### Most Commonly Tagged Characters

```python
# answers question: which characters are most represented?
# top 10 characters
character_counts = df_characters['Characters'].value_counts()
df_character_counts = character_counts.to_frame().reset_index()
df_character_counts = df_character_counts.sort_values(by='count', ascending=False)
df_character_counts[0:10]
```

```python
# plot top 10 characters
labels = {'Characters':'Character', 'count':'Number of times tagged'}
fig = px.bar(df_character_counts[0:10], x="count", y="Characters", labels=labels, title='Most Commonly Tagged Characters')
fig.show()
fig.write_html("/home/ekmys/PycharmProjects/IMT542_FinalProject_Stelter/assets/visualizations/most_common_characters.html")
```
