# Defining Exceptions (so code can rerun / try fics again if it gets nothing from Ao3)

## Learning how to define exceptions

```python
# example, from: https://www.geeksforgeeks.org/python/user-defined-exceptions-python-examples/

# Step 1: Define a custom exception class
class InvalidAgeError(Exception):
    def __init__(self, age, msg="Age must be between 0 and 120"):
        self.age = age
        self.msg = msg
        super().__init__(self.msg)

    def __str__(self):
        return f'{self.age} -> {self.msg}'
```

```python
# Step 2: Use the custom exception in your code
def set_age(age):
    if age < 0 or age > 120:
        raise InvalidAgeError(age)
    else:
        print(f"Age set to: {age}")
```

```python
# Step 3: Handling the custom exception
try:
    set_age(150)  # This will raise the custom exception
except InvalidAgeError as e:
    print(e)
```

## Defining Exception - empty list

```python
# example, from: https://www.geeksforgeeks.org/python/user-defined-exceptions-python-examples/

# Step 1: Define a custom exception class
class EmptyListError(Exception):
    def __init__(self, list_contents, msg="List is empty"):
        self.list_contents = list_contents
        self.msg = msg
        super().__init__(self.msg)

    def __str__(self):
        return f'{self.list_contents} -> {self.msg}'
```

```python
# Step 2: Use the custom exception in your code
def detect_empty_list(list_contents):
    if len(list_contents) == 0:
        raise EmptyListError(list_contents)
    else:
        print(f"List length is: {len(list_contents)}")
```

```python
# Step 3: Handling the custom exception
empty_list = []
print(len(empty_list))
print(type(empty_list))

try:
    detect_empty_list(empty_list)  # This will raise the custom exception
except EmptyListError as e:
    print(e)
```

```python
from python_retry import retry
@retry(retry_on=EmptyListError, max_retries=3, backoff_factor=1)
def get_list_length(html_list):
    list_length = len(html_list)
    return list_length
```
