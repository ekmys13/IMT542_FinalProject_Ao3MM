### Understanding Tags
- HTML tag defines different types of content in the document
- "A tag object in BeautifulSoup corresponds to an HTML or XML tag in the actual page or document."
- From tutorial: https://www.tutorialspoint.com/beautiful_soup/beautiful_soup_kinds_of_objects.htm
- Tags have a lot of methods and attributes associated with them
- Main methods / attributes: "name"

```python
soup = BeautifulSoup('<b class="boldest">TutorialsPoint</b>', 'lxml')
tag = soup.html
print (type(tag))
```

```python
# you can print the name of a tag
soup2 = BeautifulSoup('<b class="boldest">TutorialsPoint</b>', 'lxml')
tag = soup2.html
print (tag.name)
```

```python
# you can also change the name of a tag
soup3 = BeautifulSoup('<b class="boldest">TutorialsPoint</b>', 'lxml')
tag = soup3.html
tag.name = "strong"
print (tag)
```

```python
# tags have attributes that you can print out as a dictionary
soup4 = BeautifulSoup('<input type="text" name="name" value="Raju">', 'lxml')
tag = soup4.input

print (tag.attrs)
```

```python
## multi-valued attributes
# some html5 attributes have multiple values
# most common is "class" attribute which can have multiple CSS values
# other html5 attributes that have multiple values: 'rel', 'rev', 'headers', 'accesskey', 'accept-charset'
# The multi-valued attributes in beautiful soup are shown as list.
```

```python
# looking at multi-valued attributes
css_soup = BeautifulSoup('<p class="body"></p>', 'lxml')
print ("css_soup.p['class']:", css_soup.p['class'])

css_soup = BeautifulSoup('<p class="body bold"></p>', 'lxml')
print ("css_soup.p['class']:", css_soup.p['class'])
```
