```python
def create_soup_object(html_doc):
    soup = BeautifulSoup(html_doc, 'html.parser')
    return soup
```

```python
def get_soup_body(soup):
    soup_body = soup.body.prettify()
    return soup_body
```

```python
def get_all_dt(soup):
    all_dt = soup.find_all('dt')
    all_dt = list(all_dt)
    return all_dt
```

```python
def get_all_dd(soup):
    all_dd = soup.find_all('dd')
    all_dd = list(all_dd)
    return all_dd
```

```python
def get_idx_multiple_tags(dd_list):
    indices = []
    for i in range(len(dd_list)):
        if type(dd_list[i]) == list:
            indices.append(i)
    del indices[-1] # remove last index because it's a string
    return indices
```

## cleanup_multiple_tags

```python
def cleanup_multiple_tags(dd_list, indices_list):
    pattern = r"\>(.*?)\<"
    for idx in indices_list:
        current_list = dd_list[idx]
        for i in range(len(current_list)):
            match = re.search(pattern, current_list[i])
            if match:
                current_list[i] = match.group(1)
            else:
                current_list[i] = current_list[i]
            current_list[i] = current_list[i].replace("&amp;","&")
        dd_list[idx] = current_list
    return dd_list
```

```python
# function to get individual stats
from datetime import datetime

def get_publication_date(stats_str_list):
    published_str = stats_str_list[0]
    published_list = published_str.split(':')
    publication_date = published_list[1]
    return publication_date

def get_update_date(stats_str_list):
    updated_str = stats_str_list[1]
    updated_list = updated_str.split(':')
    updated_date = updated_list[1]
    return updated_date

def get_wordcount(stats_str_list):
    word_count_str = stats_str_list[2]
    word_count_list = word_count_str.split(':')
    word_count = word_count_list[1]
    word_count = word_count.replace(",","")
    word_count = int(word_count)
    return word_count

def get_chapter_count(stats_str_list):
    chapter_count_str = stats_str_list[3]
    chapter_count_list = chapter_count_str.split(':')
    chapter_count = chapter_count_list[1]
    # split string on / delimiter
    chapter_count = chapter_count.split('/')
    chapter_count = int(chapter_count[0])
    return chapter_count
```

```python
def create_fanwork_dict(dt_list, dd_list, stats_list):
    fanwork_dict = {}

    # add rating
    fanwork_dict['Rating'] = get_rating(dt_list, dd_list)

    # add warning(s)
    fanwork_dict['Archive Warnings'] = get_warnings(dt_list, dd_list)

    # category
    fanwork_dict['Category'] = get_category(dt_list, dd_list)

    # add fandom
    fanwork_dict['Fandom'] = get_fandom(dt_list, dd_list)

    # add characters
    fanwork_dict['Characters'] = get_characters(dt_list, dd_list)

    # add relationships
    fanwork_dict['Relationships'] = get_relationships(dt_list, dd_list)

    # add additional tags
    fanwork_dict['Additional Tags'] = get_additional_tags(dt_list, dd_list)

    # add language
    fanwork_dict['Language'] = get_language(dt_list, dd_list)

    # add whether it is a series
    fanwork_dict['Is Series'] = is_series(dd_list)

    # add stats
    fanwork_dict['Date Published'] = get_publication_date(stats_list)
    fanwork_dict['Date Updated'] = get_update_date(stats_list)
    fanwork_dict['Word Count'] = get_wordcount(stats_list)
    fanwork_dict['Chapter Count'] = get_chapter_count(stats_list)

    return fanwork_dict
```
