```python
import requests
import json

list_of_urls = ['www.google.com','www.apple.com','www.reddit.com']
list_of_saved_responses = [] # This creates a new list. Each request response goes here

for i in range(len(list_of_urls)):
  response = requests.get(f"http://archive.org/wayback/available?url={i}") # This appends the archive.org to each url in list_of_urls
  list_of_saved_responses.append(response[['archived_snapshots']['closest']['url']])  # This appends each JSON response to list_of_saved_responses
```
